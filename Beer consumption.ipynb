```python
# Scaling
from sklearn.preprocessing import StandardScaler
num=df.drop("Final de Semana",axis=1)
scaler=StandardScaler()
num=pd.DataFrame(scaler.fit_transform(num),columns=num.columns)
num
```

```python
num.insert(loc=0,column="Final de Semana",value=df["Final de Semana"].values)
```

```python
num.head()
```

```python
X=num.drop("Consumo de cerveja (litros)",axis=1)
y=num["Consumo de cerveja (litros)"]
```

```python
#a
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()

xtrain,xtest,ytrain,ytest=train_test_split(X,y,random_state=10,test_size=0.25)
model=linreg.fit(xtrain,ytrain)
pred=model.predict(xtest)
print("R squared",r2_score(ytest,pred))
print("mean_squared_error",mean_squared_error(ytest,pred))
print("mean_absolute_error",mean_absolute_error(ytest,pred))

# The r squred indicates that nearly 66% of variation in target variable is captured by the model. So its good
# Errors are nearly 0.25 which is decent. It means there is not much difference between actual and predicted values
```

```python
#b#Check for multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()

vif["VIF Factor"]=[variance_inflation_factor(X.values,i) for i in range (X.shape[1])]
vif["Feature"]=X.columns
print(vif)
# Since vif features are already treated above their is no multicollinearity
```

```python

#c
sns.heatmap(num.corr(),annot=True,cmap="Blues")
from statsmodels.formula.api import ols
import statsmodels.api as sm
X=sm.add_constant(X)
y=num["Consumo de cerveja (litros)"]
xtrain,xtest,ytrain,ytest=train_test_split(X,y,random_state=10,test_size=0.25)
model=sm.OLS(xtrain,ytrain).fit()
pred=model.predict(xtest)
print(model.summary())
```
